```python
import matplotlib.pyplot as plt

# Data tinggi badan siswa kelas A
tinggi_badan = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

# Menghitung Mean
mean_tinggi = sum(tinggi_badan) / len(tinggi_badan)

# Menghitung Median
sorted_tinggi = sorted(tinggi_badan)
if len(sorted_tinggi) % 2 == 0:
    median_tinggi = (sorted_tinggi[len(sorted_tinggi)//2 - 1] + sorted_tinggi[len(sorted_tinggi)//2]) / 2
else:
    median_tinggi = sorted_tinggi[len(sorted_tinggi)//2]

# Menghitung Modus
from collections import Counter
counter_tinggi = Counter(tinggi_badan)
modus_tinggi = [k for k, v in counter_tinggi.items() if v == max(counter_tinggi.values())]

# Visualisasi pusat distribusi data
plt.hist(tinggi_badan, bins=10, color='blue', edgecolor='black')
plt.axvline(mean_tinggi, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(median_tinggi, color='green', linestyle='dashed', linewidth=2, label='Median')
plt.axvline(modus_tinggi[0], color='purple', linestyle='dashed', linewidth=2, label='Modus')
plt.xlabel('Tinggi Badan')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data nilai UAS matematika kelas A
nilai_uas = [95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75]

# Menghitung Mean
mean_nilai = np.mean(nilai_uas)

# Menghitung Median
median_nilai = np.median(nilai_uas)

# Menghitung Modus
from statistics import mode
try:
    modus_nilai = mode(nilai_uas)
except:
    modus_nilai = "Tidak ada modus"

# Visualisasi pusat distribusi data
plt.hist(nilai_uas, bins=10, color='blue', edgecolor='black')
plt.axvline(mean_nilai, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(median_nilai, color='green', linestyle='dashed', linewidth=2, label='Median')
plt.xlabel('Nilai UAS Matematika')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()

# Menampilkan hasil
print("Mean: ", mean_nilai)
print("Median: ", median_nilai)
print("Modus: ", modus_nilai)
```

```python
data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]

# Mean
mean = sum(data) / len(data)
print("Mean:", mean)

# Median
sorted_data = sorted(data)
n = len(sorted_data)
median = (sorted_data[n//2] + sorted_data[(n-1)//2]) / 2
print("Median:", median)

# Range
range_data = max(data) - min(data)
print("Range:", range_data)

# Variance and Standard Deviation
variance = sum((x - mean) ** 2 for x in data) / len(data)
std_deviation = variance ** 0.5
print("Variance:", variance)
print("Standard Deviation:", std_deviation)

# Interquartile Range (IQR)
q1 = sorted_data[int(n*0.25)]
q3 = sorted_data[int(n*0.75)]
iqr = q3 - q1
print("Interquartile Range (IQR):", iqr)
```

```python
tinggi_pemain = [180, 185, 175, 190, 170, 195, 165]

# Mean
mean_tinggi = sum(tinggi_pemain) / len(tinggi_pemain)
print("Mean Tinggi:", mean_tinggi)

# Median
sorted_tinggi = sorted(tinggi_pemain)
n_tinggi = len(sorted_tinggi)
median_tinggi = (sorted_tinggi[n_tinggi//2] + sorted_tinggi[(n_tinggi-1)//2]) / 2
print("Median Tinggi:", median_tinggi)

# Range
range_tinggi = max(tinggi_pemain) - min(tinggi_pemain)
print("Range Tinggi:", range_tinggi)

# Variance and Standard Deviation
variance_tinggi = sum((x - mean_tinggi) ** 2 for x in tinggi_pemain) / len(tinggi_pemain)
std_deviation_tinggi = variance_tinggi ** 0.5
print("Variance Tinggi:", variance_tinggi)
print("Standard Deviation Tinggi:", std_deviation_tinggi)

# Interquartile Range (IQR)
q1_tinggi = sorted_tinggi[int(n_tinggi*0.25)]
q3_tinggi = sorted_tinggi[int(n_tinggi*0.75)]
iqr_tinggi = q3_tinggi - q1_tinggi
print("Interquartile Range (IQR) Tinggi:", iqr_tinggi)
```
